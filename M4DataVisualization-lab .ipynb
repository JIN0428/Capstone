## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

table_name = 'ConvertedComp'

QUERY = """
SELECT {} as ConvertedComp
FROM master
""".format(table_name)

df1 = pd.read_sql_query(QUERY,conn)
df1.plot(kind='hist')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
table_name = 'Age'

QUERY = """
SELECT {} as Age
FROM master
""".format(table_name)

df2 = pd.read_sql_query(QUERY,conn)
df2.plot(kind='box')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs
FROM master
"""

df3=pd.read_sql_query(QUERY,conn)
df3
df3.plot(kind='scatter', x='Age', y='WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age,WorkWeekHrs,CodeRevHrs
FROM master
"""

df3=pd.read_sql_query(QUERY,conn)
norm_age=(df3['Age']-df3['Age'].min())/(df3['Age'].max()-df3['Age'].min())
df3.plot(kind='scatter', x='WorkWeekHrs', y='CodeRevHrs',s=norm_age*400)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
select count(DatabaseDesireNextYear) as Count, DatabaseDesireNextYear 
from DatabaseDesireNextYear group by DatabaseDesireNextYear order by count(DatabaseDesireNextYear) DESC LIMIT 5
"""

df4=pd.read_sql_query(QUERY,conn)
df4.set_index('DatabaseDesireNextYear',inplace=True)

df4['Count'].plot(kind='pie',autopct='%1.2ff%%',startangle=90,shadow=False)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master WHERE Age BETWEEN 30 and 35
"""

df5 = pd.read_sql_query(QUERY, conn)
df5

df_median = df5.groupby(df5.Age)[['WorkWeekHrs','CodeRevHrs']].median()
df_median

df_median.plot(kind='bar')
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
Query="""
SELECT ConvertedComp,Age FROM master WHERE Age BETWEEN 45 AND 60
"""
df6=pd.read_sql_query(Query,conn)
df_median1=df6.groupby("Age")["ConvertedComp"].median()

df_median1.plot(kind="line")
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """
SELECT MainBranch,COUNT(MainBranch) as count FROM master GROUP BY MainBranch"""

df7 = pd.read_sql_query(QUERY, conn)
df7.set_index('MainBranch',inplace=True)
df7.plot(kind="barh")
```
